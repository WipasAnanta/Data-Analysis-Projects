#Importing the necessary libraries I will use for this project

```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
```

Reading in the data

```python
df =pd.read_csv('/content/share-of-global-merchandise-exports.csv')
```

Looking at the data

```python
print(df.dtypes)
```

#Data Cleaning!

Checking for missing data

```python
# checking missing in each colume
df.isna().sum()
```

Revise Data

```python
df_rev_name = df.rename(columns={'17.11.1 - Developing countries’ and least developed countries’ share of global merchandise exports (%) - TX_EXP_GBMRCH':'global_exports'})
df_rev_name
```

```python
df_rev_name["Code"].fillna("UN", inplace = True)
df_rev_name.isna().sum()
```

```python
df_export = df_rev_name
df_export
```

```python
df_2021 = df_export[(df_export['Code']== 'UN') & (df_export['Year']== 2021)]
df_sort = df_2021.sort_values('global_exports', ascending=True)
df_year = df_export[(df_export['Code']== 'UN')]
```

#Visualizing the data

```python
plt.barh(y=df_sort.Entity, width=df_sort.global_exports)
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot (x="Year", y="global_exports",
             hue="Entity",
             data=df_year)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```
